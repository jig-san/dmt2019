```python
from string import ascii_letters
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train_0.csv')
df_orig = pd.read_csv("training_set_VU_DM.csv", header=0, nrows=5000)
```

```python
df.drop(['srch_id', 'click_bool',
       'gross_bookings_usd', 'booking_bool', 'position'], axis=1, inplace=True)
df_orig.drop(['srch_id', 'click_bool',
       'gross_bookings_usd', 'booking_bool', 'position'], axis=1, inplace=True)
```

```python
b = list(set(df_orig.columns.values) - set(df.columns.values))
```

```python
a = list(set(df.columns.values) - set(df_orig.columns.values))
```

```python
sns.set(style="white")

corr = df_orig.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
